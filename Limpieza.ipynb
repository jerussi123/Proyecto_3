#### Primero se trae la data del csv

```python
import pandas as pd
import numpy as np

data_completa = pd.read_csv("datos_norte_santander.csv")
data_completa.head()
```

#### Primero sacaremos algunas columnas que no nos van a aportar en nada

```python
data = data_completa.drop(['estu_tipodocumento', 'estu_consecutivo', 'cole_cod_dane_establecimiento', 'cole_cod_dane_sede', 'cole_cod_depto_ubicacion', 
                  'cole_cod_mcpio_ubicacion', 'cole_codigo_icfes', 'cole_depto_ubicacion', 'cole_nombre_establecimiento', 'cole_nombre_sede', 
                  'cole_sede_principal', 'estu_cod_depto_presentacion', 'estu_cod_mcpio_presentacion', 'estu_cod_reside_depto', 'estu_cod_reside_mcpio', 
                  'estu_depto_presentacion', 'estu_depto_reside', 'estu_estadoinvestigacion', 'estu_estudiante', 'estu_nacionalidad', 'estu_pais_reside', 
                  'estu_privado_libertad', 'estu_fechanacimiento', 'cole_calendario'], axis=1)
data
```

#### Ahora  Revisamos que en nuestro dataframe ya no quede, ningun duplicado o null

```python
duplicates = len(data[data.duplicated()])
print(f'Numero de duplicados: {duplicates}')
missing_values = data.isnull().sum().sum()
print(f'Numero de valores nullos: {missing_values}')
```

#### Ahora revisamos cuales datos nos quedan y que tipo de columnas tienen

```python
data.dtypes
```

#### Quitamos las tildes o caracteres especiales de todos los datos

```python
data_clean = pd.DataFrame() 

for column in data.columns:
    if data[column].dtype == 'object':
        data_clean[column] = data[column].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
    else:
        data_clean[column] = data[column]
data = data_clean
data
```

```python
data.describe().transpose()
```

#### Nos damos cuenta que las unicas dos variables raras son fami_personashogar y fami_cuartoshogar, ya que pareciera estas variables se midieron de distintas maneras

```python
unicos_raros = data['fami_personashogar'].unique()
print("Datos viejos: ",unicos_raros)
mapa_agrupacion = {
    'Una': '1 a 2',
    'Dos': '1 a 2',
    '1 a 2': '1 a 2',
    'Tres': '3 a 4',
    'Cuatro': '3 a 4',
    '3 a 4': '3 a 4',
    'Cinco': '5 a 6',
    'Seis': '5 a 6',
    '5 a 6': '5 a 6',
    'Siete': '7 a 8',
    'Ocho': '7 a 8',
    '7 a 8': '7 a 8',
    'Nueve': '9 o más',
    '9 o mas': '9 o más',
    'Diez': '9 o más',
    'Once': '9 o más',
    'Doce o mas': '9 o más'
}
data['fami_personashogar'] = data['fami_personashogar'].replace(mapa_agrupacion)
nuevos_unicos = data['fami_personashogar'].unique()
print("Datos viejos: ",nuevos_unicos)
```

```python
unicos_raros = data['fami_cuartoshogar'].unique()
print("Datos viejos: ",unicos_raros)
mapa_agrupacion = {
    'Una': '1',
    'Dos': '2',
    'Tres': '3',
    'Cuatro': '4',
    'Cinco': '5',
    'Seis': '6 o mas',
    'Siete': '6 o mas',
    'Ocho': '6 o mas',
    'Nueve': '6 o más',
    'Seis o mas': '6 o más',
    'Diez o mas': '6 o más'
}
data['fami_cuartoshogar'] = data['fami_cuartoshogar'].replace(mapa_agrupacion)
nuevos_unicos = data['fami_cuartoshogar'].unique()
print("Datos viejos: ",nuevos_unicos)
```
