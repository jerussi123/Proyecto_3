# Limpieza proyecto 3

#### Primero se trae la data del csv

```python
import pandas as pd

data_completa = pd.read_csv("datos_norte_santander.csv")
data_completa.head()
```

#### Primero sacaremos algunas columnas que no nos van a aportar en nada

```python
data = data_completa.drop(['estu_tipodocumento', 'estu_consecutivo', 'cole_cod_dane_establecimiento', 'cole_cod_dane_sede', 'cole_cod_depto_ubicacion', 
                  'cole_cod_mcpio_ubicacion', 'cole_codigo_icfes', 'cole_depto_ubicacion', 'cole_nombre_establecimiento', 'cole_nombre_sede', 
                  'cole_sede_principal', 'estu_cod_depto_presentacion', 'estu_cod_mcpio_presentacion', 'estu_cod_reside_depto', 'estu_cod_reside_mcpio', 
                  'estu_depto_presentacion', 'estu_depto_reside', 'estu_estadoinvestigacion', 'estu_estudiante', 'estu_nacionalidad', 'estu_pais_reside', 
                  'estu_privado_libertad', 'estu_fechanacimiento'], axis=1)
data
```

#### Ahora revisamos los duplicados

```python
duplicates = len(data[data.duplicated()])
print(f'Number of Duplicate Entries: {duplicates}')
data[data.duplicated()]
```

#### Ahora quitamos los duplicados

```python
data.drop_duplicates(inplace=True)
data
```

#### Revisamos los valores nulos en la variable de salida puntajes globales "punt_global" y eliminamos las filas que tengan NaN o vacio

```python
count_nan = data['punt_global'].isnull().sum()
count_nan
```

```python
data = data[data['punt_global'].notna()]
data
```

#### Ahora revisamos el resto de las columnas, para revisar cual de estas tambien tiene valores en Nan o vacio

```python
missing_values = data.isnull().sum()
missing_values
```

#### Nos damos cuenta que la columna "Cole_bilingue" tiene muchos null, asi que probamos cuantas filas se eliminarian si quitamos esta columna

```python
df = data.drop('cole_bilingue', axis=1)
df = df.dropna()
df
```

#### Tambien probamos si se dejaran todas las columnas, cuantas filas quedarian al quitar todos los null

```python
data = data.dropna()
data
```

#### Escogemos el dataframe que tiene una columna menos ya que con este podemos tener 10000 datos mas

```python
data = df
data
```

#### Ahora  Revisamos que en nuestro dataframe ya no quede, ningun duplicado o null

```python
duplicates = len(data[data.duplicated()])
print(f'Numero de duplicados: {duplicates}')
missing_values = data.isnull().sum().sum()
print(f'Numero de valores nullos: {missing_values}')
```

#### Graficamos los datos para revisar los outliers

```python
import matplotlib.pyplot as plt

# Preparar el layout de los gráficos con subplots (2 filas, 5 columnas)
fig, axs = plt.subplots(2, 4, figsize=(20, 10))

# Diagramas de puntos
axs[0, 0].scatter(data.index, data['periodo'], alpha=0.5, color='blue')
axs[0, 0].set_title('Diagrama de puntos: Periodo')
axs[0, 0].set_xlabel('Índice de Observación')
axs[0, 0].set_ylabel('Periodo')

axs[0, 1].scatter(data.index, data['punt_ingles'], alpha=0.5, color='blue')
axs[0, 1].set_title('Diagrama de puntos: Puntos Ingles')
axs[0, 1].set_xlabel('Índice de Observación')
axs[0, 1].set_ylabel('Puntos Ingles')

axs[0, 2].scatter(data.index, data['punt_matematicas'], alpha=0.5, color='blue')
axs[0, 2].set_title('Diagrama de puntos: Puntos Matematicas')
axs[0, 2].set_xlabel('Índice de Observación')
axs[0, 2].set_ylabel('Puntos Matematicas')

axs[0, 3].scatter(data.index, data['punt_sociales_ciudadanas'], alpha=0.5, color='blue')
axs[0, 3].set_title('Diagrama de puntos: Puntos Sociales')
axs[0, 3].set_xlabel('Índice de Observación')
axs[0, 3].set_ylabel('Puntos Sociales')

axs[1, 0].scatter(data.index, data['punt_c_naturales'], alpha=0.5, color='blue')
axs[1, 0].set_title('Diagrama de puntos: Puntos Naturales')
axs[1, 0].set_xlabel('Índice de Observación')
axs[1, 0].set_ylabel('Puntos Naturales')

axs[1, 1].scatter(data.index, data['punt_lectura_critica'], alpha=0.5, color='blue')
axs[1, 1].set_title('Diagrama de puntos: Puntos Lectura')
axs[1, 1].set_xlabel('Índice de Observación')
axs[1, 1].set_ylabel('Puntos Lectura')

axs[1, 2].scatter(data.index, data['punt_global'], alpha=0.5, color='blue')
axs[1, 2].set_title('Diagrama de puntos: Puntos Total')
axs[1, 2].set_xlabel('Índice de Observación')
axs[1, 2].set_ylabel('Puntos Total')
```

#### Eliminamos los datos donde el puntaje sea igual a 0 en cualquiera de las variables de puntaje, Tambien si alguien saca un puntaje perfecto de 500 en el total.

```python
indexdf = data[(data['punt_ingles'] == 0) | (data['punt_matematicas'] == 0) | (data['punt_sociales_ciudadanas'] == 0) | (data['punt_c_naturales'] == 0)
                | (data['punt_lectura_critica'] == 0) | (data['punt_global'] == 0 | (data['punt_global'] == 500))].index
print(indexdf)
data.drop(indexdf, inplace=True)
data
```
